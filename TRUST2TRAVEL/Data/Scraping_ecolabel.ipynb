```python
# Importamos librerías
from bs4 import BeautifulSoup
import requests
import pandas as pd
import html
import numpy as np
import pandas as pd
from selenium import webdriver
```

```python
count = 1 #Para id_casa
lista_casas = [] #Para agregar los resultados del bucle


chromedriver = 'C:/Users/kerst/DataScience/chromedriver.exe'
options = webdriver.ChromeOptions()
options.add_argument('headless') # abre pop up para ver la navegación

driver = webdriver.Chrome(executable_path=chromedriver, options=options)

URL = 'http://ec.europa.eu/ecat/hotels-campsites/en/es/spain'
r = requests.get(URL)

soup = BeautifulSoup(r.text, 'lxml')

# Guardamos lista de casa y recorremos
casas_grid = soup.find_all(class_='col-md-4')

for casa in casas_grid:

    print('Casa {} de {}'.format(count, len(casas_grid)))

    try:
        URL_casas = 'http://ec.europa.eu' + casa.find('a')['href']
        
    except:
        categoria = None
    
    r = requests.get(URL_casas)
    soup_casa = BeautifulSoup(r.text, 'lxml')

    # Definimos id para cada casa
    id_casa = 'casa_' + str(count)

    # Nombre
    name = soup_casa.find(class_='media-heading').text
    name = name.replace('\n\r\n                                            \tEcolabel\r\n                                            \t \r\n                                            \t', '')
    name = name.replace('\r\n', '')
  
    # Info varios
    info_grid = soup_casa.find_all(class_='col-md-7 col-sm-12')

    #Direccion 
    address = info_grid[3].find_all('strong')[0].text
    address = address.replace('\n\r\n                                        \t \t\t\t\t\t\t\t\t', '')
    address = address.replace('\r\n                                        \t \t\t\t\t\t\t\t\n                         \t                                                                                  \r\n                                           \t \t\t\t\t\t\t\t', ', ')
    address = address.replace(',                                           \t \r\n                                           \t \t\t\t\t\t\t\t\r\n                                        \t \t\t\t\t\t\t\t\t', ' ')
    address = address.replace('\r\n                                        \t \t\t\t\t\t\t\t\n\n', '')

    #Telefono 
    phone = info_grid[4].find_all('strong')[0].text

    #Web
    web = info_grid[5].find('a')['href']
    
    # Imagen
    image_grid = soup_casa.find_all(class_='slides')
    image = 'http://ec.europa.eu/' + image_grid[0].find_all('img')[0]['src']
 
   # Agregamos a la lista
    lista_casas.append([
        ('id', id_casa),
        ('name', name), 
        ('address', address),
        ('phone',phone),
        ('web', web),
        ('image', image)
        ])        

    # Pasamos al siguiente id
    count += 1
```

```python
#Creamos Dataframe
df = pd.DataFrame([[x[0][1],
                    x[1][1],
                    x[2][1],
                    x[3][1],
                    x[4][1],
                    x[5][1]
                   ] for x in lista_casas],
                  columns = ['id',
                             'name',
                             'address',
                             'phone',
                             'web',
                             'image'])
```

```python
#Corregimos las URL mal escritas
df['web'] = df['web'].str.replace('http://https://','https://')
```
